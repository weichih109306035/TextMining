```python
#選擇四種類型的留言讓不同模型進行斷詞
a="泉哥 加油 祝福你事事順心如意飛鴻騰達"#含有成語
b="蚊香要查時是不是屬環境用藥....依環境用藥管理法不得廣告販售未經環保署登記核准之環境用藥，違者處刊登者新台幣6萬元以上30萬元以下罰鍰"#含有多個詞彙
c="屏東只有屏東科技大學，沒有海洋科技大學！老闆說的應該是屏東科技大學養殖系"#含有長的詞彙（例如：屏東科技大學）
d="留言:1555+數量$1555元316不鏽鋼湯火鍋*1個(34公分)本島免運 外島運費100"#含有中文詞、數字、單位、符號
```

```python
#加入list方便後續以迴圈印出斷詞後結果
sampled_data = []
sampled_data.append(a)
sampled_data.append(b)
sampled_data.append(c)
sampled_data.append(d)
sampled_data
```

```python
#使用snownlp
!pip install snownlp

import pandas as pd
import time
from snownlp import SnowNLP
```

```python
#對於每句留言各自以'/'隔開snownlp斷出的詞
snownlp_result = ['/'.join(SnowNLP(sentence).words) for sentence in sampled_data]
print(snownlp_result)
```

```python
#使用jieba
import jieba
import pandas as pd
```

```python
#對於每句留言各自以'/'隔開jieba斷出的詞
jieba_results = ['/'.join(jieba.cut(sentence)) for sentence in sampled_data]
print(jieba_results)
```

```python
#使用thulac
!pip install thulac
import thulac

#初始化THULAC對象
thu1 = thulac.thulac(seg_only=True)  # 設定模式為只進行斷詞，不進行詞性標註
```

```python
#對於每句留言各自以'/'隔開thulac斷出的詞
tulac_data = ['/'.join(thu1.cut(sentence, text=True)) for sentence in sampled_data]
print(tulac_data)
```

```python
#"泉哥 加油 祝福你事事順心如意飛鴻騰達"各自的斷詞狀況
print(snownlp_result[0])
print(jieba_results[0])
print(tulac_data[0])
```
