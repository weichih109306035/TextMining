```python
import re
```

```python
name=[]
for i in df['商品名稱']:
  name.append(re.sub(r".*\*",'',i))
print(df[:10])
print(name[:10])
```

```python
name=[]
for i in df['商品名稱']:
  name.append(re.sub(u"\(.*?\)",'',i))
print(df[:10])
print(name[40:60])
```

```python
name=[]
for i in df['商品名稱']:
  name.append(re.sub(u"\（.*?\）",'',i))
print(df[:10])
print(name[40:60])
```

```python
a=0
for value in name:
        # 检查元素是否包含数字
        if any(char=="*" for char in str(value)):
          a+=1
print(a)
```

```python
for i in range(len(name)):
  name[i]=re.sub(r'[0-9]+','',name[i])
print(df[:5])
print(name[:10])
```

```python
for i in range(len(name)):
  name[i]=re.sub(r'\*.','',name[i])
print(df[:10])
print(name[:20])
```

```python
for i in range(len(name)):
  name[i]=re.sub(r'\-.*','',name[i])
print(df[:5])
print(name[:20])
```

```python
a=0
for value in name:
        # 检查元素是否包含数字
        if any(char.isdigit() for char in str(value)):
          a+=1
print(a)
```

```python
a=0
for value in name:
        # 检查元素是否包含数字
        if any(char==" " for char in str(value)):
          a+=1
print(a)
```

```python
for i in range(len(name)):
  name[i]=name[i].replace(" ",'')
print(df[:5])
print(name[:20])
```
